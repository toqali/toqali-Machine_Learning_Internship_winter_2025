#Data Visualization

```python
df.hist(figsize=(30,15))  #Visualizing the data using histogram
plt.show()
```

```python
# Distribution of numerical features using histogram
df.hist(figsize=(12, 10), bins=20)
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()
```

```python
# Distribution of non-numerical features
for column in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y=column, order = df[column].value_counts().index)
    plt.title(f'Distribution of {column}', fontsize=16)
    plt.show()
```

```python
#numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['number'])

# prints the correlation matrix
print(numerical_df.corr())
```

```python
# Visualize Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

```python
# Boxplot for outlier detection (numerical features)
print("\nBoxplots for Outliers:")
for column in numerical_df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[column], palette="pastel")
    plt.title(f'Boxplot of {column}')
    plt.show()
```
