# 7. Visualizing Numerical Columns

We create boxplots to visualize the distribution of numerical columns and identify potential outliers

```python
# Select numerical columns
numerical_columns = data.select_dtypes(include=['number']).columns

# Define number of rows and columns for subplots
num_cols = len(numerical_columns)
num_rows = (num_cols // 3) + (num_cols % 3 > 0)  # Arrange in rows of 3

# Create subplots
fig, axes = plt.subplots(num_rows, 3, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten axes array for easy iteration

# Plot each numerical column
for i, col in enumerate(numerical_columns):
    sns.boxplot(x=data[col], ax=axes[i])
    axes[i].set_title(f"Boxplot for {col}")

# Remove empty subplots if any
for i in range(len(numerical_columns), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

We visualize the value counts of categorical columns

```python
# Select categorical columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns

# Define number of rows and columns for subplots
num_cols = len(categorical_columns)
num_rows = (num_cols // 3) + (num_cols % 3 > 0)  # Arrange in rows of 3

# Create subplots
fig, axes = plt.subplots(num_rows, 3, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten axes array for easy iteration

# Plot bar chart for each categorical column
for i, col in enumerate(categorical_columns):
    sns.countplot(x=data[col], ax=axes[i])
    axes[i].set_title(f"Bar Plot for {col}")

# Remove empty subplots if any
for i in range(len(categorical_columns), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```
