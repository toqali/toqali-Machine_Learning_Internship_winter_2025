```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())
# Drop missing values or fill them with suitable values
df.fillna(method='ffill', inplace=True)  # Example of filling missing values
```

```python
df.hist(figsize=(15, 10), bins=30, edgecolor='black')
plt.suptitle('Distributions of Numerical Features', fontsize=16)
plt.show()
```

```python
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
print(df.columns.tolist())
```

```python
# Selecting only numeric columns for visualization
numeric_features = df.select_dtypes(include=["float64", "int64"]).columns
sns.pairplot(df[numeric_features])
plt.show()
```

```python
cat_cols = ['type of meal', 'room type', 'market segment type', 'booking status']
plt.figure(figsize=(15, 10))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(2, 2, i)
    sns.countplot(data=df, x=col, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[num_cols + ['P-C', 'P-not-C']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['lead time'], kde=True, color='skyblue')
plt.title('Lead Time Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='booking status', y='average price', palette='Blues')
plt.title('Price Distribution for Different Booking Status')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='special requests', palette='Set2')
plt.title('Number of Special Requests')
plt.show()
```

```python
# Convert the column while coercing invalid dates to NaT
df['date of reservation'] = pd.to_datetime(df['date of reservation'], errors='coerce')
# Drop rows with invalid dates if necessary
df.dropna(subset=['date of reservation'], inplace=True)
# Set the date as index
df.set_index('date of reservation', inplace=True)

# Plot monthly booking trends
plt.figure(figsize=(12, 6))
df['booking status'].resample('M').count().plot()
plt.title('Monthly Booking Trends')
plt.show()
```

```python
# Scatter Plot Example
plt.figure(figsize=(8, 6))
sns.scatterplot(x="lead time", y="average price", hue="booking status", data=df)
plt.title("Lead Time vs Average Price")
plt.show()
```

```python
# Box Plot Example
plt.figure(figsize=(8, 6))
sns.boxplot(x="room type", y="average price", data=df)
plt.title("Room Type vs Average Price")
plt.xticks(rotation=45)
plt.show()

```

```python
# Stacked Bar Plot for Categorical Comparison
market_segment_counts = df.groupby(["market segment type", "booking status"]).size().unstack()
market_segment_counts.plot(kind="bar", stacked=True)
plt.title("Market Segment vs Booking Status")
plt.show()
```

```python
pivot_data = df.pivot_table(values="average price", index="room type", columns="booking status", aggfunc='mean')
sns.heatmap(pivot_data, annot=True, cmap="YlGnBu")
plt.title("Average Price by Room Type and Booking Status")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(x="room type", y="lead time", data=df, jitter=True)
plt.title("Lead Time Distribution by Room Type")
plt.xticks(rotation=45)
plt.show()
```
