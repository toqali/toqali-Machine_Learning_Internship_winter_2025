```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame 'df' is already loaded

# Filter the data into canceled and non-canceled bookings
canceled_df = df[df['booking status'] == 'Canceled']
not_canceled_df = df[df['booking status'] == 'Not_Canceled']

# Visualize the distribution of lead time for canceled vs non-canceled bookings
plt.figure(figsize=(8,5))
sns.boxplot(x='booking status', y='lead time', data=df, palette='coolwarm')
plt.title('Lead Time vs Booking Status')
plt.show()

# Calculate and compare the summary statistics for both groups
canceled_stats = canceled_df['lead time'].describe()
not_canceled_stats = not_canceled_df['lead time'].describe()

# Print the summary statistics
print("Canceled Bookings Lead Time Stats:")
print(canceled_stats)

print("\nNon-Canceled Bookings Lead Time Stats:")
print(not_canceled_stats)
```

```python
import scipy.stats as stats

# Perform the t-test to compare the lead time between canceled and non-canceled bookings
t_stat, p_value = stats.ttest_ind(canceled_df['lead time'], not_canceled_df['lead time'])

# Print the result
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation of the result
if p_value < 0.05:
    print("The difference in lead times between canceled and non-canceled bookings is statistically significant.")
else:
    print("The difference in lead times between canceled and non-canceled bookings is not statistically significant.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you already have your dataframe 'df' loaded from the CSV
# Convert 'date of reservation' column to datetime format
df['date of reservation'] = pd.to_datetime(df['date of reservation'], errors='coerce')

# Extract the month and year from the 'date of reservation'
df['month'] = df['date of reservation'].dt.month
df['year'] = df['date of reservation'].dt.year

# List of month names
month_names = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

# Create the plot
plt.figure(figsize=(10, 6))
sns.countplot(x='month', data=df, palette='viridis')

# Set the x-axis to use the month names instead of numbers
plt.xticks(ticks=range(12), labels=month_names, rotation=45)

# Set labels and title for clarity
plt.title('Distribution of Reservations by Month')
plt.xlabel('Month')
plt.ylabel('Number of Reservations')

plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression, LogisticRegression

# Load dataset (assuming df is already available)
df['booking_status_numeric'] = df['booking status'].map({'Canceled': 1, 'Not_Canceled': 0})

# Group data by Lead Time and calculate cancellation probability
lead_time_groups = df.groupby('lead time')['booking_status_numeric'].mean().reset_index()
lead_time_groups.columns = ['lead time', 'cancellation_probability']

# Plot: Probability of Cancellation vs. Lead Time
plt.figure(figsize=(8, 5))
sns.scatterplot(x=lead_time_groups['lead time'], y=lead_time_groups['cancellation_probability'], alpha=0.5)
sns.regplot(x=lead_time_groups['lead time'], y=lead_time_groups['cancellation_probability'], scatter=False, color='red')
plt.title('Probability of Cancellation vs Lead Time')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Cancellation Probability')
plt.show()

# Linear Regression Model
X_linear = df[['lead time']]
y = df['booking_status_numeric']

linear_model = LinearRegression()
linear_model.fit(X_linear, y)

# Print Linear Regression Coefficients
print(f"Linear Regression: Intercept = {linear_model.intercept_:.4f}, Coefficient = {linear_model.coef_[0]:.4f}")

# Logistic Regression Model
X_logistic = sm.add_constant(df[['lead time']])  # Add constant for intercept
logit_model = sm.Logit(y, X_logistic)
result = logit_model.fit()

# Print Logistic Regression Summary
print(result.summary())

# Plot Logistic Regression Curve
plt.figure(figsize=(8, 5))
sns.regplot(x=df['lead time'], y=df['booking_status_numeric'], logistic=True, scatter_kws={'alpha':0.3})
plt.title('Logistic Regression: Lead Time vs Cancellation Probability')
plt.xlabel('Lead Time')
plt.ylabel('Probability of Cancellation')
plt.show()
```
